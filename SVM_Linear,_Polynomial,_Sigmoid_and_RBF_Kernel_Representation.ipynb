```python
## Required python packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import svm, datasets

## load iris dataset
iris = datasets.load_iris()

## Create features and target data
x = iris.data[:, :2]
y = iris.target

## plotting
x_min, x_max = x[:,0].min() -1, x[:,0].max() +1
y_min, y_max = x[:,1].min() -1, x[:,1].max() +1
h= (x_max/x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max,h),np.arange(y_min, y_max,h))
x_plot = np.c_[xx.ravel(), yy.ravel()]
```

```python
y
```

```python
# lineaer  kernel:
svc_linear = svm.SVC(kernel = "linear", C=5.0)
svc_linear.fit(x,y)

# z: y_pred
z = svc_linear.predict(x_plot)
z = z.reshape(xx.shape)

# plotting:
plt.figure(figsize = (8,5))
plt.contour(xx, yy, z, alpha = 0.3)
plt.scatter(x[:,0], x[:,1], c=y)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.title("SVM with linear kernal")
```

```python
# polynomial kernel:
svm_poly= svm.SVC(kernel = "poly", C=5.0)
svm_poly.fit(x,y)

# z: y_pred
z = svm_poly.predict(x_plot)
z = z.reshape(xx.shape)

# plotting:
plt.figure(figsize = (15,5))
plt.subplot(121)
plt.contour(xx, yy, z, alpha = 0.3)
plt.set_cmap("gist_rainbow")
plt.scatter(x[:,0], x[:,1], c=y, cmap = plt.cm.Set1)
plt.xlabel("Sepal length")
plt.xlim(xx.min(), xx.max())
plt.ylabel("Sepal width")
plt.title("SVM with polynomial kernal")
plt.plot()
```

```python
# Sigmoid kernel:
svm_sigmoid = svm.SVC(kernel = "sigmoid", C=1.0)
svm_sigmoid.fit(x,y)

# z: y_pred
z = svm_sigmoid.predict(x_plot)
z = z.reshape(xx.shape)

# plotting:
plt.figure(figsize = (8,5))
plt.contour(xx, yy, z, alpha = 0.3)
plt.scatter(x[:,0], x[:,1], c=y)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.title("SVM with igmoid kernal")
```

```python
# RBF kernel:
svm_rbf= svm.SVC(kernel = "rbf", C=5.0)
svm_rbf.fit(x,y)

# z: y_pred
z = svm_rbf.predict(x_plot)
z = z.reshape(xx.shape)

# plotting:
plt.figure(figsize = (15,5))
plt.subplot(121)
plt.contour(xx, yy, z, alpha = 0.3)
plt.set_cmap("gist_rainbow")
plt.scatter(x[:,0], x[:,1], c=y, cmap = plt.cm.Set1)
plt.xlabel("Sepal length")
plt.xlim(xx.min(), xx.max())
plt.ylabel("Sepal width")
plt.title("SVM with RBF kernal")
plt.plot()
```
